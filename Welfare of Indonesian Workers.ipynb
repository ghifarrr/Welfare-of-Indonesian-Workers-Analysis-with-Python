### Garis Kemiskinan Per-kapita

Garis Kemiskinan merupakan representasi dari rupiah yang diperlukan atau harga yang dibayarkan agar penduduk dapat hidup layak secara minimum yang mencakup pemenuhan kebutuhan minimum makanan (setara dengan 2.100 kilokalori per kapita per hari) dan non makanan essential.

Garis Kemiskinan (GK) diperoleh dari hasil survei (sampel). Angka kemiskinan yang dirilis BPS merupakan data makro dan merupakan hasil Susenas (Survey Sosial Ekonomi Nasional) yang menunjukkan persentase penduduk miskin terhadap jumlah penduduk dalam suatu wilayah.

### Rata-rata Pengeluaran Per-kapita

Rata-rata pengeluaran perkapita adalah ukuran yang menggambarkan jumlah uang yang dihabiskan oleh penduduk suatu negara dalam suatu periode tertentu, umumnya diukur per tahun. Ini adalah salah satu indikator penting dalam menganalisis tingkat kesejahteraan ekonomi suatu negara dan standar hidup penduduknya.
Rata-rata pengeluaran perkapita mencerminkan tingkat kesejahteraan ekonomi suatu negara. Semakin tinggi rata-rata pengeluaran perkapita, semakin tinggi standar hidup di negara tersebut

Rata-rata pengeluaran per-kapita diperoleh dari hasil lembaga statistik nasional (BPS) ataupun survei rumah tangga (sampel). Angka rata-rata pengeluaran per-kapita yang dirilis BPS merupakan pengumpulan data terhadap responden di berbagai rumah tangga dipilih secara acak untuk berpartisipasi dalam survei. Mereka diminta memberikan informasi terperinci tentang pengeluaran mereka selama periode tertentu, yang kemudian diolah dan dihitung untuk mendapatkan rata-rata pengeluaran perkapita.

### Rata-rata Upah Pekerja Per-jam

Rata-rata upah pekerja per-jam adalah rata-rata imbalan finansial yang diberikan kepada pekerja sebagai balas jasa atas pekerjaan yang mereka lakukan berdasarkan 1 jam kerja yang dihabiskan. Upah pekerja perjam dihitung dengan membagi jumlah upah yang diterima oleh pekerja dalam suatu periode dengan jumlah jam kerja yang dilakukan. Pemerintah dapat menggunakan informasi tentang upah pekerja perjam untuk menetapkan kebijakan upah minimum, mengatasi disparitas upah, dan meningkatkan kesejahteraan pekerja.

Beberapa negara menetapkan upah minimum, yang merupakan jumlah upah terendah yang diizinkan untuk dibayar kepada pekerja. Upah minimum bertujuan untuk melindungi pekerja dengan menetapkan standar minimum yang harus dipatuhi oleh pengusaha.

Rata-rata upah pekerja per-jam diperoleh dari hasil lembaga statistik nasional (BPS) ataupun laporan pengusaha. Angka rata-rata upah pekerja per-jam yang dirilis diperoleh dengan mewawancarai pekerja atau meminta pengusaha untuk melaporkan informasi tentang upah yang dibayarkan kepada pekerja per jam. Survei dapat mencakup berbagai sektor ekonomi dan tingkat pendidikan atau keterampilan pekerja.

### Upah Minimum Provinsi (UMP)

Upah minimum provinsi adalah standar upah minimum yang ditetapkan oleh pemerintah di tingkat provinsi atau daerah administratif setingkat provinsi di suatu negara. Upah minimum provinsi berlaku di tingkat provinsi dan dapat berbeda antar provinsi dalam satu negara.

Tujuan diberlakukannya upah minimum provinsi karena UMP adalah instrumen kebijakan yang penting untuk meningkatkan kondisi pekerja dan memastikan bahwa upah yang diterima oleh pekerja mencukupi untuk kebutuhan hidup di tingkat lokal. Penetapan upah minimum provinsi mempertimbangkan kondisi ekonomi dan sosial setempat untuk mencapai keseimbangan yang baik antara hak pekerja dan keberlanjutan ekonomi

Kita akan membuat dataframe baru bernama gk_tren dimana gk_tren hanya mengambil nilai-nilai kolom pada 'provinsi' = INDONESIA, dikarenakan tren yang akan kita analisis adalah tren pada data seluruh INDONESIA bukan setiap provinsi-nya.

#### Apakah terdapat perbedaan signifikan antara ambang batas garis kemiskinan di daerah perkotaan, perdesaan, dan perdesaanperkotaan?

Kita dapat mengetahui perbedaan signifikan antara ambang batas garis kemiskinan didaerah perkotaan, perdesaan, perdesaanperkotaan setiap tahunnya dari tahun 2013 sampai tahun 2022 dengan cara membuat heatmap agar kita dapat mengetahui perbedaan data satu sama lain setiap tahunnya pada setiap jenis daerah dengan representasi visual dengan pengelompokkan warna.
```python
gk
```

```python
#Convert kolom 'tahun' ke format datetime
gk_tren['tahun'] = pd.to_datetime(gk['tahun'], format='%Y') 

# Siapkan data untuk plot
grouped = gk_tren.groupby(['daerah', 'tahun']).agg({'gk': ['mean']})
grouped.columns = ['Rata-rata Garis Kemiskinan Perkapita']

# Membuat heatmap dari data
plt.figure(figsize=(12, 8))
sns.heatmap(data=grouped, cmap='coolwarm', linewidths=1, annot=True, fmt='.2f')
plt.title('Rata-rata Nilai Garis Kemiskinan Berdasarkan Jenis Daerah')
plt.ylabel('Jenis Daerah')
plt.show()
```

#### Provinsi mana yang memiliki garis kemiskinan perkapita tertinggi dan terendah?

Kita dapat mengetahui provinsi mana yang memiliki ambang batas garis kemiskinan perkapita tertinggi dan terendah sepanjang tahun 2013 sampai tahun 2022 bisa kita cari dengan menggunakan metode max dan min pada seluruh data gk disetiap provinsi dengan mengecualikan INDONESIA. Nilai max dan mix yang dimaksud adalah nilai maksimal ambang batas garis kemiskinan pada setiap provinsi dan nilai minimal ambang batas garis kemiskinan pada setiap provinsi dengan mengabaikan jenis daerah mana yang ditempati, selanjutnya nilai tersebut kita visualisasikan dengan stacked bar chart.

```python
gk
```

```python
# Membuat tabel 'gk_sorted' tanpa provinsi = INDONEISA
gk_sorted = gk[gk['provinsi'] != 'INDONESIA']

# Tentukan max_gk dan min_gk untuk setiap provinsi
max_gk = gk_sorted.groupby('provinsi')['gk'].max().reset_index()
min_gk = gk_sorted.groupby('provinsi')['gk'].min().reset_index()

# Buat DataFrame untuk menyimpan posisi dan label untuk nilai gk maksimum
pos_max_gk = pd.DataFrame({'provinsi': max_gk['provinsi'],
                           'gk': max_gk['gk'],
                           'label': max_gk['gk'].apply(lambda x: f'{x:.2f}')})

# Buat DataFrame untuk menyimpan posisi dan label untuk nilai gk minimum
pos_min_gk = pd.DataFrame({'provinsi': min_gk['provinsi'],
                           'gk': min_gk['gk'],
                           'label': min_gk['gk'].apply(lambda x: f'{x:.2f}')})

# Membuat plot
fig, ax = plt.subplots(figsize=(16, 9))

# Membuat sebuah bar chart untuk garis kemiskinan perkapita maksimum
ax.barh(pos_max_gk['provinsi'], pos_max_gk['gk'], color='red', label='Maximum gk')

# Membuat sebuah bar chart untuk garis kemiskinan perkapita minimum
ax.barh(pos_min_gk['provinsi'], pos_min_gk['gk'], color='blue', label='Minimum gk')

# Tambahkan nilai maksimum gk ke kiri bar chart
for i, value in enumerate(pos_max_gk['label']):
    ax.text(pos_max_gk['gk'][i] + 0.1, i, value, color='black', fontweight='bold', horizontalalignment='left')

# Tambahkan nilai minimum gk ke kiri bar chart
for i, value in enumerate(pos_min_gk['label']):
    ax.text(pos_min_gk['gk'][i] - 0.1, i, value, color='black', fontweight='bold', horizontalalignment='left')

# Atur Label dan judul untuk plot
ax.set_xlabel('Garis Kemiskinan')
ax.set_ylabel('Provinsi')
ax.set_title('Minimum and Maksimum Garis Kemiskinan per kapita setiap Provinsi di Tahun 2013-2022')

# Putar label x-axis dari horizontal ke vertikal
plt.xticks(rotation=90)

# Tampilkan plot (chart)
plt.legend()
plt.show()
```

**Dengan adanya outlier pada analisa pertama dan kedua pada DataFrame Upah Minimum Provinsi, kita harus mengetahui dan mencari dimanakah titik outlier sebelum melakukan kembali analisisnya. Jika titik outlier adalah data yang tidak valid (ada kesalahan entry data) maka akan dilakukan perbaikan data.**

```python
# Atur limit
Q1 = ump[ump['tahun'].between(2018, 2022)]['ump'].quantile(0.25)
Q3 = ump[ump['tahun'].between(2018, 2022)]['ump'].quantile(0.75)

IQR = Q3 - Q1

#1,5 adalah pengganda yang digunakan untuk menentukan batas atas dan bawah untuk pencilan. 
#Pengali ini berasal dari definisi rentang interkuartil (IQR), yang merupakan ukuran penyebaran statistik atau dengan kata lain, seberapa tersebarnya nilai dalam kumpulan data.
lower_limit = Q1 - 1.5 * IQR

upper_limit = Q3 + 1.5 * IQR

# Mencari outlier berdasarkan tahun 2018 sampai tahun 2022
outliers = ump[(ump['tahun'].between(2018, 2022)) & ((ump['ump'] < lower_limit) | (ump['ump'] > upper_limit))]
```

Telah dibuktikan berdasarkan kesimpulan diatas bahwa data ump provinsi DKI Jakarta pada tahun 2021 berada pada data ump provinsi Gorontalo pada tahun 2021, begitu juga data ump provinsi Gorontalo pada tahun 2021 berada pada data ump provinsi Jambi pada tahun 2021.

Maka hal yang harus dilakukan sekarang adalah melakukan perbaikan data pada DataFrame menggunakan metode Data Imputation, yaitu dengan mengganti data yang tidak valid pada data ump beberapa provinsi pada tahun 2021. Data ump beberapa provinsi pada tahun 2021 yang tidak valid yaitu data ump provinsi Banten pada tahun 2021 sampai dengan data ump provinsi Kepulauan Bangka Belitung pada tahun 2021.

```python
# List provinsi yang akan diubah (dari BENGKULU sampai KEP. BANGKA BELITUNG)
provinsi_to_change = ['BANTEN', 'BENGKULU', 'DI YOGYAKARTA', 'DKI JAKARTA', 'GORONTALO',
                      'JAMBI', 'JAWA BARAT', 'JAWA TENGAH', 'JAWA TIMUR',
                      'KALIMANTAN BARAT', 'KALIMANTAN SELATAN', 'KALIMANTAN TENGAH', 'KALIMANTAN TIMUR', 'KALIMANTAN UTARA',
                      'KEP. BANGKA BELITUNG']

# Mengganti nilai 'ump' pada setiap provinsi sesuai dengan urutan
for i in range(len(provinsi_to_change) - 1):
    current_provinsi = provinsi_to_change[i]
    next_provinsi = provinsi_to_change[i + 1]
    
    # Mengambil nilai ump provinsi saat ini dan tahun 2021
    nilai_baru = ump.loc[(ump['provinsi'] == next_provinsi) & (ump['tahun'] == 2021), 'ump'].values[0]
    
    # Mengganti nilai 'ump' pada provinsi berikutnya dengan nilai baru
    ump.loc[(ump['provinsi'] == current_provinsi) & (ump['tahun'] == 2021), 'ump'] = nilai_baru
    ump = ump[ump['provinsi'] != 'INDONESIA'].sort_values(by=['provinsi','tahun'])

# Mengganti nilai 'ump' pada 'provinsi' KEP BANGKA BELITUNG di 'tahun' 2021
ump.loc[(ump['provinsi'] == 'KEP. BANGKA BELITUNG') & (ump['tahun'] == 2021), 'ump'] = 3230023.0

# Menampilkan DataFrame setelah perubahan
print((ump).to_string())
```

Setelah dilakukan perbaikan data dengan mengganti nilai UMP pada provinsi Banten pada tahun 2021 dengan nilai yang sesuai dari provinsi berikutnya, dan melanjutkan perubahan nilai UMP hingga mencapai provinsi Kepulauan Bangka Belitung (metode Data Imputation), maka dapat disimpulkan bahwa data Series UMP yang telah diperbaiki adalah data UMP sebenarnya.
